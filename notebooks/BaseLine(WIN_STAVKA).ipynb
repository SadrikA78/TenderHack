```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.34)
```

```python
#preds_model_lgb.loc[preds_model_lgb < 0] = 0
```

```python
models = [linear_model.LinearRegression(),
          linear_model.Ridge(alpha=.5),
          linear_model.Lasso(alpha=0.1),
          linear_model.LassoLars(alpha=.1),
          linear_model.BayesianRidge(),
          #ARDRegression(compute_score=True),
          #svm.SVR(),
          GradientBoostingRegressor(),]

for model in models:
    model_name = model.__class__.__name__
    model.fit(X_train, y_train)
    result = model.predict(X_test)
    rmse = sqrt(mean_squared_error(result, y_test))
    print(model_name, rmse)
    
```

```python
ereg = VotingRegressor(estimators=[('gb', linear_model.LinearRegression()), ('rf', linear_model.Ridge(alpha=.5)), ('lr', linear_model.BayesianRidge())])
ereg = ereg.fit(X_train, y_train)
```

```python
result = ereg.predict(X_test)
result = pd.Series(result)
result.loc[result < 0] = 0
rmse = sqrt(mean_squared_error(result, y_test))
```
